# Linear regression for predicting lattice parameters from STEM-HAADF images

Import modules

```python
import matplotlib.pyplot as plt
import numpy as np
import math
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error
```

Load all data points of 'fcc111' for the distances of descriptor (1st column) and the lattice constants (2nd column) in training set

```python
with open('data_points_fcc111.dat', 'r') as f:
    matrix = [[float(num) for num in line.split(' ')] for line in f]

matrix = np.array(matrix)
X = np.array(matrix[:,0])
X = np.reshape(X,(X.shape[0],1))
y = np.array(matrix[:,1])
y = y.transpose()
```

Construct the support vector regression model using linear kernel

```python
regressor = SVR(kernel='linear')#, C=100, gamma='auto')
regressor.fit(X,y)
print(regressor)
```

Show the relation between the distances of descriptor and the lattice constants

```python
plt.figure(figsize = (5, 4), dpi=100)
plt.scatter(X, y, color = 'blue')
plt.plot(X, regressor.predict(X), color = 'red')
plt.title('fcc111')
plt.xlabel('Descriptor (Angstrom)')
plt.ylabel('Lattice constant (Angstrom)')
plt.show()
```

Measure mse(mean squared error) and rmse (root mean squared error)

```python
mse=mean_squared_error(y, regressor.predict(X))
print("Mean Squared Error: %.2f" %(mse))

rmse = math.sqrt(mse)
print("Root Mean Squared Error: %.2f" %(rmse))
```
