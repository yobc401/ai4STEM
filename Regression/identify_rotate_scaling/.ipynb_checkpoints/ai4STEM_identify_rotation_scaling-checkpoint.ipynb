Import modules

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import os
import matplotlib.cm as cm
from matplotlib.pyplot import figure, show, axes, sci
from matplotlib import cm, colors
from matplotlib.font_manager import FontProperties
from numpy import amin, amax, ravel
from matplotlib.colors import LinearSegmentedColormap
import ai4STEM_function

filename = "HAADF.tif" #Target STEM-HAADF image

hyperparameters = 0
num_X, num_Y = ai4STEM_function.segment(filename, hyperparameters) #Segment STEM-HAADF image
```

Learn rotated angles and lattice constants of all local window inside STEM-HAADF image. For predicting lattice constants, put the name of the class of crystal structure and lattice plane (e.g., fcc111, fcc110, fcc100, fcc211, bcc111, bcc110, bcc100, hcp0001, hcp10m10, hcp1m1m20, hcp2m1m10).

```python
matrix_rotation = np.zeros((num_X,num_Y))
matrix_scaling = np.zeros((num_X,num_Y))

for x in range(num_X):
    for y in range(num_Y):
        filename = 'local_images_' + str(x+1) + '_' + str(y+1) + '.png'

        angle, dist = ai4STEM_function.learn(filename)
        lattice_const = ai4STEM_function.map_from_dist_to_latConst('fcc111',dist)
        print('local_image_%d_%d -> Rotated angle: %.2f, Lattice constant: %.2f (Dist.: %.2f)' %(x+1,y+1,angle,lattice_const,dist))
        
        matrix_rotation[x,y] = angle
        matrix_scaling[x,y] = lattice_const
```

Plot the identification map for rotated angles

```python
cmap='Blues'

fig = plt.figure()
plt.imshow(matrix_rotation, interpolation='none', cmap=cmap)
#plt.title(title)
cb=plt.colorbar()
cb.ax.tick_params(labelsize=20)
plt.tight_layout()
# add this otherwise the x-axis gets cut
#plt.gcf().subplots_adjust(bottom=0.25)
#plt.ylabel(title_true_label)
#plt.xlabel(title_pred_label)
plt.show()
plt.axis('off')
#plt.savefig('colormap_rotate.png',bbox_inches='tight',pad_inches=0)
plt.clf()
```

Plot the identification map for lattice constants

```python
cmap='copper'

fig = plt.figure()
plt.imshow(matrix_scaling, interpolation='none', cmap=cmap)
#plt.title(title)
cb=plt.colorbar()
cb.ax.tick_params(labelsize=20)
plt.tight_layout()
# add this otherwise the x-axis gets cut
#plt.gcf().subplots_adjust(bottom=0.25)
#plt.ylabel(title_true_label)
#plt.xlabel(title_pred_label)
plt.show()
plt.axis('off')
#plt.savefig('colormap_scale.png',bbox_inches='tight',pad_inches=0)
plt.clf()
```
